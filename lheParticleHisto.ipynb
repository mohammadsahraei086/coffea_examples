```python
import awkward as ak
import numpy as np
import hist
from coffea.nanoevents.methods import vector
from coffea.nanoevents import NanoEventsFactory, DelphesSchema, NanoAODSchema, BaseSchema

import matplotlib.pyplot as plt
```

```python
################################################################################################################
#Photon#
################################################################################################################
```

```python
def get_data(file , particleID):
    events = NanoEventsFactory.from_root(file, schemaclass=BaseSchema).events()
    
    mask = (events['Particle/Particle.PID'] == particleID) & (events['Particle/Particle.PT']>=10) & (events['Particle/Particle.Eta'] < 5)
    particle_PT = ak.flatten(events['Particle/Particle.PT'][mask]).compute()
    particle_Eta = ak.flatten(events['Particle/Particle.Eta'][mask]).compute()
    particle_Phi = ak.flatten(events['Particle/Particle.Phi'][mask]).compute()
    particle_E = ak.flatten(events['Particle/Particle.E'][mask]).compute()
    return particle_PT, particle_Eta, particle_Phi, particle_E
```

```python
photon_PT1000, photon_Eta1000, photon_Phi1000, photon_E1000 = get_data("unweighted_events1.root", 22)
photon_PT500, photon_Eta500, photon_Phi500, photon_E500 = get_data("unweighted_events2.root", 22)
photon_PT1200, photon_Eta1200, photon_Phi1200, photon_E1200 = get_data("unweighted_events3.root", 22)
```

```python
plt.figure(figsize=(12, 12))
plt.suptitle("Photon")
#plt.subplots_adjust(wspace=0.5, hspace=0.5)


plt.subplot(2, 2, 1)
plt.hist(photon_PT500, range=[0,600], bins=50, label='T_Mass=500')
plt.hist(photon_PT1000, range=[0,600], bins=50, label='T_Mass=1000')
plt.hist(photon_PT1200, range=[0,600], bins=50, label='T_Mass=1200')
plt.xlabel("Photon_PT")
plt.ylabel("Number of Events")
#plt.title("Photon_PT")
plt.legend()

plt.subplot(2, 2, 2)
plt.hist(photon_Eta500, range=[-5,5], bins=50, label='T_Mass=500', fill=False, color='blue', histtype='step')
plt.hist(photon_Eta1000, range=[-5,5], bins=50, label='T_Mass=1000', fill=False, color='orange', histtype='step')
plt.hist(photon_Eta1200, range=[-5,5], bins=50, label='T_Mass=1200', fill=False, color='green', histtype='step')
plt.xlabel("Photon_Eta")
plt.ylabel("Number of Events")
#plt.title("Photon_Eta")
plt.legend()

plt.subplot(2, 2, 3)
plt.hist(photon_Phi500, range=[-3.5,3.5], bins=50, label='T_Mass=500', fill=False, color='blue', histtype='step')
plt.hist(photon_Phi1000, range=[-3.5,3.5], bins=50, label='T_Mass=1000', fill=False, color='orange', histtype='step')
plt.hist(photon_Phi1200, range=[-3.5,3.5], bins=50, label='T_Mass=1200', fill=False, color='green', histtype='step')
plt.xlabel("Photon_Phi")
plt.ylabel("Number of Events")
#plt.title("Photon_Eta")
plt.legend()

plt.subplot(2, 2, 4)
plt.hist(photon_E500, range=[0, 1200], bins=50, label='T_Mass=500', fill=False, color='blue', histtype='step')
plt.hist(photon_E1000, range=[0, 1200], bins=50, label='T_Mass=1000', fill=False, color='orange', histtype='step')
plt.hist(photon_E1200, range=[0, 1200], bins=50, label='T_Mass=1200', fill=False, color='green', histtype='step')
plt.xlabel("Photon_E")
plt.ylabel("Number of Events")
#plt.title("Photon_Eta")
plt.legend()


plt.savefig('Photon.jpg', format='jpg')
```

```python
def get_data(file):
    events = NanoEventsFactory.from_root(file, schemaclass=BaseSchema).events()
    events['Particle/Particle.PID'] = events['Particle/Particle.PID'][:,3:]
    events['Particle/Particle.PT'] = events['Particle/Particle.PT'][:,3:]
    events['Particle/Particle.Eta'] = events['Particle/Particle.Eta'][:,3:]
    events['Particle/Particle.Phi'] = events['Particle/Particle.Phi'][:,3:]
    events['Particle/Particle.E'] = events['Particle/Particle.E'][:,3:]
    
    mask = ((events['Particle/Particle.PID'] == 1) | (events['Particle/Particle.PID'] == 2) | (events['Particle/Particle.PID'] == 3) |\
            (events['Particle/Particle.PID'] == 4) | (events['Particle/Particle.PID'] == -1) | (events['Particle/Particle.PID'] == -2) |\
            (events['Particle/Particle.PID'] == -3) | (events['Particle/Particle.PID'] == -4)) &\
            (events['Particle/Particle.PT']>=10) & (events['Particle/Particle.Eta'] < 5)
    particle_PT = ak.flatten(events['Particle/Particle.PT'][mask]).compute()
    particle_Eta = ak.flatten(events['Particle/Particle.Eta'][mask]).compute()
    particle_Phi = ak.flatten(events['Particle/Particle.Phi'][mask]).compute()
    particle_E = ak.flatten(events['Particle/Particle.E'][mask]).compute()
    return particle_PT, particle_Eta, particle_Phi, particle_E
```
