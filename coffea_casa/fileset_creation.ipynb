```python
ddc = DataDiscoveryCLI()
ddc.do_blocklist_sites(sites='T3_CH_CERN_OpenData')
fileset = ddc.load_dataset_definition(dataset_definition,
                           query_results_strategy="all",
                           replicas_strategy="round-robin");
```

```python
import copy

for key in fileset.keys():
    list = copy.deepcopy(fileset[str(key)]['files'])
    for file in list:
        changed_file = file.replace(file[file.find("root://")+7 : file.find("/store/")], 'xcache/')
        print(changed_file)
        fileset[str(key)]['files'][changed_file] = fileset[str(key)]['files'].pop(file)
print(fileset[str(key)]['files'])
```
